```python
# Dependencies and Setup
import pandas as pd
import requests
import numpy as np
import glob
import datetime as dt 
from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.plotting import gmap, save, output_file
from bokeh.models import GMapOptions
from bokeh.models import ColumnDataSource
from bokeh.io import export_png
from PIL import Image
import matplotlib.pyplot as plt

# Import API key
from config import gkey
```

```python
# Combine Daily FIRMS reports into one csv
path = r"C:\Users\Sam\Desktop\D.A\Ukraine_War\FIRMS\*.csv"
fnames = []
df = pd.DataFrame()
for fname in glob.glob(path):
    #fnames.append(fname)
    data = pd.read_csv(fname)
    df = pd.concat([df,data], axis = 0)
```

```python
# Get locations, Dates
df["acq_time"] = pd.to_datetime(df["acq_time"])

locations = df[["latitude", "longitude"]]
date = df[["acq_date"]]
brightness = df[["brightness"]]

# info box template
#info_box_template = """
#<dl>
#<dt>Latitude</dt><dd>{latitude}</dd>
#<dt>Longitude</dt><dd>{longitude}</dd>
#<dt>Brightness</dt><dd>{brightness}</dd>
#</dl>
#"""

#FIRMS_info = [info_box_template.format(**row) for index, row in df.iterrows()]
```

```python
# Kyiv
df1 = df.loc[df['acq_date'].astype(str) == '2022-02-23']
df1['brightnessadjusted'] = df1['brightness'] - 200
df1['brightnessadjusted'] = (df1['brightnessadjusted'] / df1['brightnessadjusted'].abs().max())/1.5
output_file("0223.html")
bokeh_width, bokeh_height = 500,400
lat, lng = 51, 30.5234
def plot(lat, lng, zoom=10, map_type='roadmap'):
    gmap_options = GMapOptions(lat = lat, lng = lng, 
                               map_type=map_type, zoom=7)
    p = gmap(gkey, gmap_options, title='Battle of Kyiv on 2022-02-23', 
             width=bokeh_width, height=bokeh_height)
    source = ColumnDataSource(df1)
    center = p.circle('longitude', 'latitude', size=5, alpha='brightnessadjusted', 
                      color='red', source=source)
    save(p)
    export_png(p, filename = "0223.png")
    return p
p = plot(lat, lng)
```

```python
geolocator = Nominatim(user_agent="http")
```

```python
Latitude = "50"
Longitude = "30"
 
location = geolocator.reverse(Latitude+","+Longitude)
address = location.raw['address']

# Display
print(location)

print(address['state'])
```

```python
def reverse_geocoding(lat, lon):
        try:
            location = geolocator.reverse(f'{lat},{lon}')
            address = location.raw['address']
            #print(address['state'])
            return address['state']
        except:
            return None
```
